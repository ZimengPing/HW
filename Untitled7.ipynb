```python
import turtle
import math

def edge(x, y):
    turtle.penup()
    turtle.goto(x - 190, y + 100)  
    turtle.pendown()
    turtle.begin_fill()
    turtle.fillcolor('red')
    turtle.pen(pencolor='red')
    for i in range(2):
        turtle.forward(380)
        turtle.right(90)
        turtle.forward(200)
        turtle.right(90)
    turtle.end_fill()  
    turtle.penup()

    for i in range(6):
        turtle.goto(-190, 100 - 200 / 13 * (2 * i + 1))
        turtle.forward(0)

        turtle.begin_fill()  
        turtle.fillcolor('white')  
        turtle.pen(pencolor='white')
        for j in range(2):
            turtle.forward(381)
            turtle.right(90)
            turtle.forward(200 / 13)
            turtle.right(90)
        turtle.end_fill()  

    turtle.penup()

    turtle.goto(-190, 100)
    turtle.setheading(0)
    turtle.pendown()
    turtle.begin_fill()
    turtle.fillcolor('blue')
    turtle.pen(pencolor='blue')
    for i in range(2):
        turtle.forward(380 / 5 * 2)
        turtle.right(90)
        turtle.forward(199 / 13 * 7) 
        turtle.right(90)
    turtle.end_fill()  
    turtle.penup()


def star():
    for i in range(9):
        if i % 2 == 0:
            for j in range(6):
                turtle.penup()
                turtle.goto(-190 + 380 / 5 * 2 / 12 * (2 * j + 1), 100 - (200 / 13 * 7) / 10 * (i + 1))
                turtle.setheading(90)
                turtle.forward(12.32 / 2 * math.cos(18 * math.pi / 180))
                turtle.left(162)
                turtle.pendown()
                turtle.pen(pencolor='white')

                turtle.begin_fill()
                turtle.fillcolor('white')
                for n in range(5):
                    turtle.forward(12.32 * math.cos(18 * math.pi / 180))
                    turtle.left(144)

                turtle.end_fill()  
        else:
            for j in range(5):
                turtle.penup()
                turtle.goto(-190 + 380 / 5 * 2 / 12 * (2 * j + 2), 100 - (200 / 13 * 7) / 10 * (i + 1))
                turtle.setheading(90)
                turtle.forward(12.32 / 2 * math.cos(18 * math.pi / 180))
                turtle.left(162)
                turtle.pendown()
                turtle.pen(pencolor='white')

                turtle.begin_fill()
                turtle.fillcolor('white')
                for n in range(5):
                    turtle.forward(12.32 * math.cos(18 * math.pi / 180))
                    turtle.left(144)

                turtle.end_fill()  


if __name__ == '__main__':
    turtle.setup(width=468, height=312)  
    turtle.hideturtle()
    turtle.pen(speed=10)
    edge(0, 0)
    star()

turtle.done()
```
